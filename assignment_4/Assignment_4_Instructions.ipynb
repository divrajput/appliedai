## GETTING RANDOM NUMBER FOR K-VALUE IN KNN FROM UNIFORM DISTRIBUTION

```python
def unique_values_fn(param_range,num):
    #param: tuple to store lowest and highest value of uniform distribution (both are positive)
    #num: it stores number of k-values to be stored 
    (min_val,max_val)= param_range
    params= {}
    lst= set() # set to store values of k such that it do not duplicates
    while len(lst) <= num:
        a= np.random.randint(1, high= max_val, size=1) # To generate random integer from uniform dist.
        lst.add(a[0])
    params['n_neighbors']=list(lst) # storing k-values in dictionary
    return params
```

## FUNCTION TO IMPLEMENT RANDOM SEARCH

```python
from sklearn.metrics import accuracy_score
def RandomSearch(x_train,y_train,classifier, params, folds):
    #classifier: KNN model
    #params: random values of K for KNN
    #folds: for cross validation
    #print(x_train.shape)
    trainscores = []  # to store training score for different k-values
    testscores  = []  # to store cross validation score for different k-values
    for k in tqdm(params['n_neighbors']):
        trainscores_folds = []
        testscores_folds  = []
        num= len(x_train)//folds  # variable to store number that will divide _train into equal parts
        for fold in range(folds):
            ####### Making training and cross validation data ###########
            if fold==0:
                cv_x= x_train[0: num]
                train_x= x_train[num:]
                cv_y= y_train[0: num]
                train_y= y_train[num:]
            elif fold == folds-1:
                cv_x= x_train[fold*num :]
                train_x= x_train[0:fold*num]
                cv_y= y_train[fold*num :]
                train_y= y_train[0:fold*num]
            else:
                cv_x= x_train[fold*num : fold*num+num]
                train_x= np.append(x_train[0:fold*num], x_train[fold*num+num:],axis=0)

                cv_y= y_train[fold*num : fold*num+num]
                train_y= np.append(y_train[0:fold*num], y_train[fold*num+num:],axis=0)
        
            ##############################################################
                
            classifier.n_neighbors = k
            classifier.fit(train_x,train_y)

            Y_predicted = classifier.predict(cv_x)
            testscores_folds.append(accuracy_score(cv_y, Y_predicted))

            Y_predicted = classifier.predict(train_x)
            trainscores_folds.append(accuracy_score(train_y, Y_predicted))
        trainscores.append(np.mean(np.array(trainscores_folds)))
        testscores.append(np.mean(np.array(testscores_folds)))
    return trainscores,testscores
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import random
import warnings
warnings.filterwarnings("ignore")


neigh = KNeighborsClassifier()

(min_val,max_val) = (1,50)    # tuple for min and max values of uniform dist.
num = 10                      # value for number of k-values for KNN
params = unique_values_fn((min_val,max_val),num)   # Getting num random values for K
folds = 10       ## NUMBER OF FOLDS FOR CROSS VALIDATION

trainscores,testscores = RandomSearch(X_train, y_train, neigh, params, folds)

#### dictionary to store both train score and vaildation score so that i can sorted them in increasing order of params####
scores = {}
for i,param in enumerate(params['n_neighbors']):
    scores[param] = [trainscores[i], testscores[i]]
scores = dict(sorted(scores.items()))
###########################################################################################################################

plt.plot(scores.keys(),[item[0] for item in list(scores.values())], label='train cruve')
plt.plot(scores.keys(),[item[1] for item in list(scores.values())], label='validation curve')
plt.title('Hyper-parameter VS accuracy plot')
plt.legend()
plt.show()
```

We can see difference between train and validation accuracy is minimum for k=45 So there is very small overfitting present.
Moreover the validation accuracy is maximum at k=45.

## FUNCTION FOR PLOTTING DECISION BOUNDARY

```python
def plot_decision_boundary(X1, X2, y, clf):
        # Create color maps
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    x_min, x_max = X1.min() - 1, X1.max() + 1
    y_min, y_max = X2.min() - 1, X2.max() + 1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
    # Plot also the training points
    plt.scatter(X1, X2, c=y, cmap=cmap_bold)
    
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("2-Class classification (k = %i)" % (clf.n_neighbors))
    plt.show()
```
